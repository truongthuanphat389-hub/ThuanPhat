```python
#14.Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
#a.Tạo mảng a chứa các số nguyên ngẫu nhiên.

import random
def tao_mang(m,n):
 return[[random.randint(1,20) for _ in range(n)]for _ in range(m)]

def in_mang(a):
  for row in a:
    print(*row)

#b.Xuất các phần tử thuộc dòng k.
def Xuat_dong(a,k):
  print("Dòng",k,":",a[k])

#c.Xuất các phần tử thuộc cột k.
def Xuat_cot(a,k):
  print("Cột",k,":", end=" ")
  for row in a:
     print(row[k], end=" ")
  print()


#d.Tìm dòng có tổng lớn nhất 45.
def Dong_co_tong_lon_nhat_duoi_45(a):
    max_sum = -1
    index = -1

    for i in range(len(a)):
        s = sum(a[i])
        if s < 45 and s > max_sum:
            max_sum = s
            index = i

    if index == -1:
        return None
    return index, max_sum

#e.Tìm cột có tích nhỏ nhất.
def Cot_tich_nho_nhat(a):
    num_rows = len(a)
    num_cols = len(a[0])
    min_product = float('inf')
    min_product_col_index = -1
    for j in range(num_cols): # Iterate through columns
        product = 1
        for i in range(num_rows): # Iterate through rows
            product *= a[i][j]
        if product < min_product:
            min_product = product
            min_product_col_index = j
    return min_product_col_index, min_product

#f.Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
def Xuat_dong_chan_cot_le(a):
    result_elements = []
    for i in range(0, len(a), 2):
        for j in range(1, len(a[0]), 2):
            result_elements.append(str(a[i][j]))
    return " ".join(result_elements)

#g.Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
def TBC_chan_dong_le(a):
    tong = 0
    dem = 0

    for i in range(len(a)): # Iterate through rows
        if i % 2 == 0:  # dòng lẻ (nếu tính từ 1, index 0,2,4... là dòng 1,3,5...)
            for j in range(len(a[i])):
                if a[i][j] % 2 == 0:
                    tong += a[i][j]
                    dem += 1

    if dem == 0:
        return 0
    return tong / dem

#h.Tính trung bình cộng các phần tử thuộc biên.
def TBC_bien(a):
    m, n = len(a), len(a[0])
    tong = dem = 0
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1
    return tong/dem if dem else 0

#i.Tính trung bình tích các phần tử không thuộc biên.
def TBC_khong_bien(a):
    m, n = len(a), len(a[0])
    tich = 1
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tich *= a[i][j]
            dem += 1

    return tich/dem if dem else 0


    # =====  TEST =====

m=int(input("Nhập m:"))
n=int(input("Nhập n:"))

a=tao_mang(m,n)
print("Mảng:")

in_mang(a)
k = int(input("Nhập k:"))

Xuat_dong(a, k)
Xuat_cot(a, k)
kq = Dong_co_tong_lon_nhat_duoi_45(a)

if kq is not None:
    print("Dòng:", kq[0], "tổng:", kq[1])
else:
    print("Không có dòng nào có tổng < 45")
print("Dòng có tổng <45 lớn nhất:",Dong_co_tong_lon_nhat_duoi_45(a))
print("Cột có tích nhỏ nhất:", Cot_tich_nho_nhat(a))
print("Phần tử dòng chẵn cột lẻ:", Xuat_dong_chan_cot_le(a))
print("TBC chẵn dòng lẻ:", TBC_chan_dong_le(a))
print("TBC biên:", TBC_bien(a))
print("TBC tích không biên:", TBC_khong_bien(a))
```

```python
#15.Một sinh viên gồm các thông tin sau:
#Mã sinh viên là chuỗi có 10 ký tự
#Tên là chuỗi tối đa 20 ký tự
#Năm sinh là một số nguyên
#Điểm trung bình là một số thực a.

#a.Xây dựng cấu trúc SINHVIEN mô tả một sinh viên
class SinhVien:
  def __init__(self, msv, ten, namsinh, diembb):
        self.msv = msv
        self.ten = ten
        self.namsinh = namsinh
        self.dtb = diembb

#b.Cho một mảng có n sinh viên.
def nhap_ds_sinh_vien(n):
    ds = []
    for i in range(n):
        print(f"\nNhập thông tin sinh viên thứ {i+1}:")
        while True:
            ma = input("Nhập mã sinh viên ( 10 ký tự): ")
            if len(ma) == 10:
                break
            else:
                print("Mã sinh viên không được vượt quá 10 ký tự. Vui lòng nhập lại.")
        while True:
            ten = input("Nhập tên sinh viên (tối đa 20 ký tự): ")
            if len(ten) <= 20:
                break
            else:
                print("Tên sinh viên không được vượt quá 20 ký tự. Vui lòng nhập lại.")
        while True:
            try:
                nam = int(input("Nhập năm sinh: "))
                break
            except ValueError:
                print("Năm sinh phải là một số nguyên. Vui lòng nhập lại.")
        while True:
            try:
                dtb = float(input("Nhập điểm trung bình: "))
                break
            except ValueError:
                print("Điểm trung bình phải là một số thực. Vui lòng nhập lại.")
        ds.append(SinhVien(ma, ten, nam, dtb))
    return ds

#c.Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp, biết rằng sinh viên đủ điều kiện lên lớp khi điểm trung bình lớn hơn hoặc bằng 5.
def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.dtb >= 5)

#d.Xuất các sinh viên đủ 20 tuổi.
def sv_20_tuoi(ds):
    nam_hien_tai = 2026 # Giả sử năm hiện tại là 2026
    found = False
    for sv in ds:
        if nam_hien_tai - sv.namsinh >= 20:
            print(sv.ten)
            found = True
    if not found:
        print("Không có sinh viên nào đủ 20 tuổi.")

#e.Đếm số sinh viên học hệ đại học, biết rằng sinh viên hệ DH có mã sinh viên chứa 2 ký tự DH ở vị trí 2,3 trong chuỗi. VD: 02DH0001.
def dem_he_DH(ds):
    return sum(1 for sv in ds if len(sv.msv) >= 4 and sv.msv[2:4] == "DH")

#f.Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»
def dem_ten_lan(ds):
    return sum(1 for sv in ds if  sv.ten.split()[-1]=="Lan")

#g.Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»
def dem_ho_phan(ds):
    return sum(1 for sv in ds if sv.ten.split()[0] == "Phan")

# =====  TEST =====
n = int(input("Nhập số lượng sinh viên: "))
ds_sinh_vien = nhap_ds_sinh_vien(n)

print("\n--- KẾT QUẢ --- ")
print(f"Số sinh viên đủ điều kiện lên lớp: {dem_len_lop(ds_sinh_vien)}")
print(f"Số sinh viên hệ Đại học: {dem_he_DH(ds_sinh_vien)}")
print(f"Số sinh viên có tên 'Lan': {dem_ten_lan(ds_sinh_vien)}")
print(f"Số sinh viên có họ 'Phan': {dem_ho_phan(ds_sinh_vien)}")
print("Sinh viên đủ 20 tuổi:")
sv_20_tuoi(ds_sinh_vien)
```

```python
# BTVN
#1.Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là min của 2 phần tương ứng từ 2 mảng cho trước. Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
#Ví dụ:
#Mảng a: 3 9 1 4
#Mảng b: 2 7 4 3 2 8
#Mảng kết quả: 2 7 1 3 2 8
def tron_min(a, b):
    c = []
    m, n = len(a), len(b)
    # Explicitly use the built-in min function to avoid shadowing by a variable named 'min'
    for i in range(__builtins__.min(m, n)):
        c.append(__builtins__.min(a[i], b[i]))

    if m > n:
        c.extend(a[n:])
    else:
        c.extend(b[m:])

    return c
a = [3,9,1,4]
b = [2,7,4,3,2,8]
print(tron_min(a,b))
```

```python
#2.Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
import random

def tao_ma_tran_random(m, n):
    a = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(-10, 10))  # số từ -10 → 10
        a.append(row)
    return a
#a.Tính tổng các phần tử thuộc tam giác trên của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n.
def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:  # trên + chéo phụ
                tong += a[i][j]
    return tong

#b.Chuyển các phần tử âm thành trị tuyệt đối của nó trong a.
def doi_so_am(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a

#c.Thay các phần tử chẵn trong a bằng số nguyên x cho trước.
def thay_chan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

#d.Kiểm tra a có toàn chẵn không?
def toan_chan(a):
    for row in a:
        for x in row:
            if x % 2 != 0:
                return False
    return True

#e.Cho ma trận vuông a cấp n, viết hàm kiểm tra a có đối xứng không. Biết rằng ma trận đối xứng là ma trận có a[i][j] = a[j][i].
def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True

#f.Kiểm tra ma trận vuông a cấp n có đường chéo chính tăng dần không?
def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

#g.Xuất các phần tử thuộc tam giác dưới của đường chéo phụ kể cả đường chéo phụ trong ma trận vuông a cấp n.
def tam_giac_duoi_cheo_phu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()

#h.Kiểm tra ma trận vuông a cấp n có đường chéo phụ giảm dần không?
def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-2-i]:
            return False
    return True

def nhap_ma_tran():
    m = int(input("Nhap so dong m: "))
    n = int(input("Nhap so cot n: "))

    a = []
    for i in range(m):
        row = []
        for j in range(n):
            x = int(input(f"a[{i}][{j}] = "))
            row.append(x)
        a.append(row)

    return a

m = int(input("Nhap cap ma tran n: "))
a = tao_ma_tran_random(m, m)
# Tạo ma trận
# a = tao_ma_tran_random(m, n) # This line was causing the NameError

# In ra
print("Ma trận:")
for row in a:
    print(row)
# a
print("Tong tam giác trên chéo phụ:", tong_tam_giac_tren_cheo_phu(a))

# b
print("Sau khi đổi số âm:")
print(doi_so_am([row[:] for row in a]))  # copy để không làm đổi a gốc

# c
print("Thay số chẵn = 0:")
print(thay_chan([row[:] for row in a], 0))

# d
print("Toàn chẵn:", toan_chan(a))

# e
print("Đối xứng:", doi_xung(a))

# f
print("Chéo chính tăng dần:", cheo_chinh_tang(a))

# g
print("Tam giác dưới chéo phụ:")
tam_giac_duoi_cheo_phu(a)

# h
print("Chéo phụ giảm:", cheo_phu_giam(a))
```
